```python
# Initialize the Client for API v2
client = tweepy.Client(bearer_token)

# The query for searching (You can still use hashtags like #FreePalestine)
# The search query format is mostly the same, but the method is different.
search_query = "#FreePalestine lang:id -is:retweet" # Example: search for the hashtag, in English, and exclude retweets
```

```python
try:
    # Use the V2 endpoint method: client.search_recent_tweets()
    # Free access allows max_results up to 100 per request.
    response = client.search_recent_tweets(
        query=search_query,
        max_results=100
        # Optional: You can add more fields if needed, e.g., tweet_fields=["created_at", "public_metrics"]
    )

    # V2 search results are in response.data
    tweets = response.data

    if tweets:
        print(f"Found {len(tweets)} tweets:")
        for tweet in tweets:
            # Note: tweet object properties might differ slightly from v1.1
            print(f"- Tweet ID: {tweet.id}: {tweet.text}")
    else:
        print("Found 0 tweets.")

except tweepy.TweepyException as e:
    print(f"Error during tweet search: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
if tweets:
    with open("hasil_crawling.txt", "w", encoding="utf-8") as f:
        for t in tweets:
            f.write(t.text.replace("\n", " ") + "\n")
    print("File berhasil disimpan sebagai hasil_crawling.txt")
else:
    print("Tidak ada tweet untuk disimpan.")
```
